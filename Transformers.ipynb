```python
from transformers import pipeline

translation = pipeline("translation_en_to_fr",model="Helsinki-NLP/opus-mt-en-fr")

prompt = """Tanslate English to Spanish:

English: Good Morning

spanish: Buen dia

English: Thank you

spanish:                """

print(translation(prompt))
```
